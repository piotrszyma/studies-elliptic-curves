```python
import csv
import dataclasses
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

INPUT_PATH = 'pollard_rho_affine_20_out.csv'
```

```python


@dataclasses.dataclass
class Row:
    x: int
    y: int
    type_: str


rows = []

with open(INPUT_PATH, newline="") as csvfile:
    raw_rows = csv.reader(csvfile, delimiter=";")

    for row in raw_rows:
        try:
            type_, x, y = row
        except ValueError:
            continue

        x = int(x)
        y = int(y)
        rows.append(Row(type_=type_, x=x, y=y))

max_x = max(row.x for row in rows)
max_y = max(row.y for row in rows)
```

```python

plt.scatter([row.x for row in rows], [row.y for row in rows], s=2)
plt.show()
```
